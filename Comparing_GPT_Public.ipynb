```python
def predict(model, tokenizer, context, query):
    # Constructing the prompt with clear separators and instructions
    input_text = f"Context: {context}\nQuestion: {query}\nAnswer:"
    inputs = tokenizer.encode_plus(input_text, return_tensors='pt', padding='longest', truncation=True)

    # Generate a response
    outputs = model.generate(
        inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        num_beams=5,
        max_new_tokens=50,  # Allows for generating up to 50 new tokens
        length_penalty=1.0,
        early_stopping=True,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode and return the answer, removing the constructed prompt from the beginning of the generated text
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    prompt_length = len(tokenizer.encode(input_text, add_special_tokens=False))
    answer = answer[prompt_length:].strip()

    return answer
```

```python
def normalize_text(s):
  """Removing articles and punctuation, and standardizing whitespace are all typical text processing steps."""
  import string, re

  def remove_articles(text):
    regex = re.compile(r"\b(a|an|the)\b", re.UNICODE)
    return re.sub(regex, " ", text)

  def white_space_fix(text):
    return " ".join(text.split())

  def remove_punc(text):
    exclude = set(string.punctuation)
    return "".join(ch for ch in text if ch not in exclude)

  def lower(text):
    return text.lower()

  return white_space_fix(remove_articles(remove_punc(lower(s))))
```

```python
def compute_exact_match(prediction, truth):
    return int(normalize_text(prediction) == normalize_text(truth))
```

```python
def compute_f1(prediction, truth):
  pred_tokens = normalize_text(prediction).split()
  truth_tokens = normalize_text(truth).split()

  # if either the prediction or the truth is no-answer then f1 = 1 if they agree, 0 otherwise
  if len(pred_tokens) == 0 or len(truth_tokens) == 0:
    return int(pred_tokens == truth_tokens)

  common_tokens = set(pred_tokens) & set(truth_tokens)

  # if there are no common tokens then f1 = 0
  if len(common_tokens) == 0:
    return 0

  prec = len(common_tokens) / len(pred_tokens)
  rec = len(common_tokens) / len(truth_tokens)

  return 2 * (prec * rec) / (prec + rec)
```

```python
def give_an_answer(model_name, model, tokenizer, context, query, answer):
    prediction = predict(model, tokenizer, context, query)
    em_score = compute_exact_match(prediction, answer)
    f1_score = compute_f1(prediction, answer)

    result = {
        "Model": model_name,
        "Question": query,
        "Prediction": prediction,
        "True Answer": answer,
        "EM": em_score,
        "F1": f1_score
    }

    print(f"#### {model_name} ####")
    print(f"Question: {query}")
    print(f"Prediction: {prediction}")
    print(f"True Answer: {answer}")
    print(f"EM: {em_score}")
    print(f"F1: {f1_score}")
    print("\n")

    return result
```

```python
import requests
from collections import Counter

# Helper function to calculate precision and recall
def precision_recall(reference, candidate):
    reference_tokens = reference.split()
    candidate_tokens = candidate.split()
    common = Counter(reference_tokens) & Counter(candidate_tokens)
    num_same = sum(common.values())
    if num_same == 0:
        return 0, 0
    precision = num_same / len(candidate_tokens)
    recall = num_same / len(reference_tokens)
    return precision, recall

# Function to calculate F1 score
def calculate_f1_score(reference, candidate):
    precision, recall = precision_recall(reference, candidate)
    if precision + recall == 0:
        return 0
    f1_score = 2 * (precision * recall) / (precision + recall)
    return f1_score

# Function to query GPT-3.5-turbo and get the answer
def query_gpt_3_5_turbo(api_key, context, question):
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json",
    }

    data = {
        "model": "gpt-3.5-turbo",
        "messages": [{"role": "user", "content": f"{context}\n{question}"}]
    }

    response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=data)
    return response

# Function to compare the model's answer with the expected answer
def compare_answers(model_answer, expected_answer):
    # Process the model's answer (e.g., extract the relevant part if necessary)
    processed_answer = model_answer.strip()

    # Calculate metrics like Exact Match or F1 score
    em_score = int(processed_answer == expected_answer)
    f1_score = calculate_f1_score(processed_answer, expected_answer)

    return em_score, f1_score

# Your context and question
context = "Hi! My name is Alexa and I am 21 years old. I used to live in Peristeri of Athens, but now I moved on in Kaisariani of Athens."
question = "Where does Alexa live now?"
expected_answer = "Kaisariani of Athens"

# Your API key for GPT-3.5-turbo
api_key = "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"

# Make the API call to get the answer from GPT-3.5-turbo
response = query_gpt_3_5_turbo(api_key, context, question)

# Check if the response is successful
if response.status_code == 200:
    # Extract the answer from the response
    response_data = response.json()

    # Since the structure is different, we access 'message' directly instead of 'messages'
    model_answer = response_data["choices"][0]["message"]["content"] if "choices" in response_data and "message" in response_data["choices"][0] else None

    # Check if model_answer is not None
    if model_answer:
        # Compare the model's answer with the expected answer
        em_score, f1_score = compare_answers(model_answer, expected_answer)

        # Print the results
        print(f"Question: {question}")
        print(f"Model Answer: {model_answer}")
        print(f"Expected Answer: {expected_answer}")
        print(f"Exact Match Score: {em_score}")
        print(f"F1 Score: {f1_score}")
    else:
        print("No model answer was found in the response.")
else:
    print(f"Failed to get a response from the API, status code: {response.status_code}")
    print(response.text)  # Print the error response from the API
```
